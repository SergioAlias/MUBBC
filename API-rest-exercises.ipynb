1. Write a program to list the description of all the human populations included in the Ensembl database. 

````python
import requests

server = "https://rest.ensembl.org/"
species = "homo_sapiens"
endpoint = f"info/variation/populations/{species}"

data = requests.get(server+endpoint, headers = {'content-type':'application/json'})

data_json = data.json()

for pop in data_json:
  print(pop["description"])
````

2. Write a program to obtain the DNA sequence for the human gene with stable ID ENST00000288602, and to save the sequence in a file called 'ENST00000288602.txt'

````python
import requests

server = "https://rest.ensembl.org/"
gene_id = "ENST00000288602"
endpoint = f"/sequence/id/{gene_id}"

data = requests.get(server+endpoint, headers = {'content-type':'text/plain'})

with open (f"{gene_id}.txt", "w") as f:
  f.write(data.text)
````

3. Write a program to read the contents of the previous file, and search for all the subsequences of 'T' base of length 4. For each one, it must print its position from the beginning of the sequence. For example:

```
POS: 352
POS: 353
POS: 438
...
```

A sequence of length five must be considered two times. For example, if a subsequence of lenght 5 starts at position 50, it should considered as two subsenqueces of length 4 starting at positions 50 and 51.

````python
with open (f"{gene_id}.txt", "r") as f:
  seq = f.read()

for i in range(0, len(seq)-5):
  if seq[i:i+4] == "TTTT":
    print(f"POS: {i+1}")
````
